```python
import pandas
```

```python
df = pandas.DataFrame([[1,"a"],[2,"b"],[3,"c"],[7,"d"],[10,"e"],[11,"f"]], columns=["Time", "Data"])
df
```

```python
df.Time.diff().fillna(0)
```

```python
label = iter(range(100))
grp = next(label)

df["Group"] = 0

for i in range(len(df) - 1):
    if (df.iloc[i].Time - df.iloc[i+1].Time) > -2:
        df.loc[i+1, "Group"] = grp
    else:
        grp = next(label)
        df.loc[i+1, "Group"] = grp
df
```

```python
df = pandas.DataFrame([[1,"a"],[2,"b"],[3,"c"],[7,"d"],[10,"e"],[11,"f"]], columns=["Time", "Data"])
df
```

```python
df[['t_1', 'd_1']] = df.shift(-1)
df
```

```python
df['cluster'] = ((df['t_1'] - df['Time'] > 2))
df
```

```python
df['cluster'] = df['cluster'].shift(1).fillna(False).astype(int).cumsum()
df
```
